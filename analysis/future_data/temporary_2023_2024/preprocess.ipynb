```python
import pandas as pd
import ast
import re

# Load the data
file_path = 'analysis/future_data/temporary_2023_2024/es_future_final_chronos_2023_2024.csv'
df = pd.read_csv(file_path)

def clean_array_string(value):
    try:
        if pd.isna(value):  # Handle NaN values
            return value
        # Extract exactly 3 floats inside brackets with flexible spacing
        match = re.match(r'^\[\s*([\d\.]+)\s+([\d\.]+)\s+([\d\.]+)\s*\]$', value.strip())
        if match:
            corrected_value = f"[{match.group(1)}, {match.group(2)}, {match.group(3)}]"
            print("it is ok: ", ast.literal_eval(corrected_value))
            return corrected_value
        else:
            print("Regex doesn't work for this value: ", value)
        return value  # Leave non-matching values unchanged
    except Exception as e:
        print(f"Error in value: {value} - {e}")
        return None  # Or 'Invalid' if you prefer

# Apply the function to the 'Result' column
df['Result'] = df['Result'].apply(clean_array_string)

# # Save the updated dataframe
df.to_csv(file_path, index=False)

print("Preprocessing complete. All corrected values are safely converted.")
```
